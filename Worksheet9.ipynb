# 3. Write a function that takes an argument (n) and calculates n!.

```python
def factorial (n):
  fact=1
  for i in range(n,1,-1):
    fact=fact*i
  return fact

n=6
print("The factorial of factorial {} is {}.".format(n,factorial(n)))
```
